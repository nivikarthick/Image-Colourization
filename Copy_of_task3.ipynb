```python
import cv2
import numpy as np

img = cv2.imread("test.jpg")

# Check if the image was loaded successfully
if img is None:
    print("Error: Could not load image 'test.jpg'. Please ensure the file exists in the correct directory.")
else:
    mask = np.zeros(img.shape[:2], np.uint8)

    mask[50:200,50:200] = 255   # Example region

    img[mask==255] = [255,0,0]

    cv2.imwrite("conditional.jpg", img)
```
